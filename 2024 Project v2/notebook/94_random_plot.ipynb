```python

import numpy as np
import matplotlib.pyplot as plt
import math

# define the function
def f(x):
    x = 1-x
    y = +x+1/(1-x)
    return y


x = np.linspace(0.01,1, 100)

# create an array of y values by applying the function to x
y = np.array([f(val) for val in x])  # Apply the function element-wise

# plot the function but on the x-axis show only the value of 0 and in the place of 5 show "output"

plt.plot(x, y)
plt.xlabel('Emissions')
plt.ylabel('Abatement Cost')
plt.xticks([0, 1], ['0', 'output'])
plt.title('2 Definition of Abatement')
# Add a vertical line on x=5 to show the discontinuity red color dashed
plt.axvline(x=1, color='r', linestyle='--')
#plt.grid(True)
plt.show()
```
